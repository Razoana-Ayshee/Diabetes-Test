```python
data=pd.read_csv("diabetess.csv")
data
```

```python
x=data[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
```

```python
y=data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
#x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10) 
#"random_state=10" for the smaple will be same all the time
```

```python
from keras.models import Sequential
from keras.layers import Dense
neu_net = Sequential()
neu_net.add(Dense(8,activation='relu',input_shape=(8, )))
neu_net.add(Dense(8,activation='relu'))#relu works better for hidden layers
neu_net.add(Dense(8,activation='relu'))
neu_net.add(Dense(8,activation='relu'))
neu_net.add(Dense(8,activation='relu'))
neu_net.add(Dense(8,activation='relu'))
neu_net.add(Dense(1,activation='sigmoid'))#our project is binary classification so used sigmoid

```

```python
#neu_net.summary()
```

```python
neu_net.compile(optimizer='sgd',
              loss='mean_squared_error',
             metrics=['accuracy'])
```

```python
neu_net.fit(x_train, y_train, epochs=1200)
```

```python
y_pred =neu_net.predict(x_test)
print(y_pred)
```

```python
test_loss, test_acc = (neu_net.evaluate(x_test, y_test, verbose=2))
test_acc=test_acc*100
test_acc
```
